# R($\phi$) gate
La compuerta R($\phi$) realiza una rotación de un ángulo $\phi$ alrededor del eje-z, donde $\phi$ es un número real.
$$R(\phi) = \left(\begin{array}{cc} 1 & 0 \\ 0 & e^{i\phi} \end{array}\right)$$

```python
from qiskit import *
from qiskit.visualization import *

import math

pi=math.pi

qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.rz(pi/2,1)
qc.draw(output='mpl')

simulador = Aer.get_backend('statevector_simulator')
resultados = execute(qc, backend = simulador).result()
vector_estado = resultados.get_statevector()
plot_bloch_multivector(vector_estado)
```
