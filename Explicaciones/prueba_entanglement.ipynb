```python
# Importo todo de qiskit
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector, plot_histogram

# Creo un circuito cuantico de 2 cubits
qc = QuantumCircuit(6)
# Apply Hadamard gate to create superposition on the first set of 3 qubits
qc.h(range(3))

# Apply Hadamard gate to create superposition on the second set of 3 qubits
qc.h(range(3, 6))

# Apply entangling gates between corresponding qubits
for i in range(3):
    qc.cx(i, i + 3)

qc.draw('mpl')
```

```python
# Measure all qubits
qc.measure_all()

# Choose a backend (simulator or real quantum device)
backend = Aer.get_backend('qasm_simulator')

# Execute the quantum circuit on the chosen backend
job = execute(qc, backend, shots=1024)

# Get the result
result = job.result()
counts = result.get_counts()

# Print the measurement results
print("Measurement results:", counts)
```

```python
plot_histogram(counts)
```

```python
# Creamos un circuito cuántico con cuatro qubits
qc = QuantumCircuit(4)

# Aplicamos puertas Hadamard a los dos primeros qubits (a)
qc.h(0)
qc.h(1)

# Aplicamos puertas CNOT para entrelazar los qubits correspondientes entre (a) y (b)
qc.cx(0, 2)
qc.cx(1, 3)

# Visualizamos el circuito
qc.draw('mpl')
```
