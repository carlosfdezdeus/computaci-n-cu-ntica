```python
# Importo las librerías necesarios
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
# CREACIÓN DEL CIRCUITO CUÁNTICO:
# Circuito cuántico de 1 cubit y 1 bit
qc = QuantumCircuit(1,1)
    
# Realizamos una puerta x sobre el cubit
qc.x(0)
```

```python
# REPRESENTACIÓN VECTOR DE ESTADO:
# Importo el simulador 'statevector_simulator' de Qiskit:
#     statevector (vector de estado), describe el estado cuántico de nuestro cúbit
simulador = Aer.get_backend('statevector_simulator')

#Ejecutamos el circuito y cogemos sus resultados:
resultados = execute(qc, backend = simulador).result()

# Calculo vector de estado (statevector):
vector_estado = resultados.get_statevector()

# Imprimimos el vector de estado para ver que está funcionando
print(vector_estado)

# Para poder ver nuestros dibujos
%matplotlib inline      

#Dibujamos el circuito cuántico
qc.draw(output='mpl')
```

```python
# Imprimo el vector de estado en una Esfera de Bloch:
plot_bloch_multivector(vector_estado)
```

```python
# Agrego un medidor al circuito: mido el cubit 0 y lo pongo en el bit 0
qc.measure([0], [0])
```

```python
# Importo el simulador de Qiskit:
simulador = Aer.get_backend('qasm_simulator')
```

```python
#Ejecutamos el circuito 1024 veces y cogemos sus resultados:
resultados = execute(qc, backend = simulador, shots = 1024).result()
```

```python
# Cogemos las cuentas del circuito:
cuentas = resultados.get_counts()
```

```python
# Importamos las herramientas de visualización de qiskit
from qiskit.tools.visualization import plot_histogram

# Mostramos el resultado del circuito
plot_histogram(cuentas)
     # El 100% de las veces se convierte al estado 1
```

```python
# CREACIÓN DEL CIRCUITO CUÁNTICO:
# Circuito cuántico de 1 cubit y 1 bit
qc = QuantumCircuit(1,1)

# Realizamos una puerta x sobre el cubit
qc.x(0)
```

```python
# REPRESENTACIÓN MATRICIAL:
# Importo el simulador 'unitary_simulator' de Qiskit:
#     matriz_unitaria, describe el estado cuántico de nuestro cúbit en formato matricial
simulador = Aer.get_backend('aer_simulator_unitary')
qc = transpile(qc, simulador)

# Run and get counts
resultados = simulador.run(qc).result()
matriz_unitaria = resultados.get_unitary(experiment=None, decimals=None)
print(matriz_unitaria)
# Para poder ver nuestros dibujos
%matplotlib inline      

#Dibujamos el circuito cuántico
qc.draw(output='mpl')
```

```python
# REPRESENTACIÓN MATRICIAL:
# Importo el simulador 'unitary_simulator' de Qiskit:
#     matriz_unitaria, describe el estado cuántico de nuestro cúbit en formato matricial
simulador = Aer.get_backend('aer_simulator_unitary')

#Ejecutamos el circuito y cogemos sus resultados:
resultados = execute(qc, simulador).result()

# Calculo la matriz unitaria (unitary):
matriz_unitaria = resultados.get_unitary()

# Imprimimos la matriz unitaria para ver que está funcionando
print(matriz_unitaria)

# Para poder ver nuestros dibujos
%matplotlib inline      

#Dibujamos el circuito cuántico
qc.draw(output='mpl')
```
