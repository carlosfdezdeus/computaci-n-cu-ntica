```python
# Importo todo de qiskit
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector

# Creo un circuito cuantico de 3 cubits y 3 bits
qc = QuantumCircuit(7)
qc.x(1)
qc.y(2)
qc.z(3)
# Estado de superposiciÃ³n estoy al 50% en |0> y |1>
qc.h(4) 
qc.h(5)
qc.z(5)
qc.h(6)
qc.s(6)
# Dibujo el circuito
%matplotlib inline
qc.draw(output='mpl')
```

```python
simulador = Aer.get_backend('statevector_simulator')
resultados = execute(qc, backend = simulador).result()
vector_estado = resultados.get_statevector()
plot_bloch_multivector(vector_estado)
```
