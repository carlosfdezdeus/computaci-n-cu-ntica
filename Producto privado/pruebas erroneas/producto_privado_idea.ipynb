### Producto privado
$
\left |\varphi_{00} \right> = \sum_{i=0}^{p-1} \left|i\right> \left|i\right>
$ 

$
[X(x_{a}^A)Z(z_{a}^A)) \otimes (X(x_{b}^B)Z(z_{b}^B)]\left |\varphi_{00} \right>
$
