### Generalización a dimensión 'n'



En este caso se han definido únicamente 4 encodings posibles para la generalización a dimensiones arbitrarias. En realididad la cantidad de encodigns posibles

Es necesario mencionar que los parametros: $\alpha y \beta$ son raices priviticvas $mod p$ donde p a de ser un número primo.

$$2(n-1)^2 + 2(n - 1)$$

Para dimensiones muy grandes como $pn2^{40}$, la cantidad de encodings aumentaría notablemente. Por otro lado, este tipo de encondings con dimensioens tan grandes no se podría aplicar al ámbito clásico, ya que las matrices no entrarían en memoria mientras que en el ámbito de la cuántica es abitual trabajar con matrices grandes.
